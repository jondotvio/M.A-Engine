# Transcript stats

Quick exploration of a day's transcripts. Counts turns, average length, and which speaker tends to break consensus first.


```python
import json
from pathlib import Path
from collections import Counter

day = Path('../data/transcripts').glob('*/*.json')
speakers = Counter()
for path in day:
    data = json.loads(path.read_text())
    for t in data['turns']:
        speakers[t['speaker']] += 1

print(speakers)
```
